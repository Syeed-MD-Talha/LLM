```python
from sentence_transformers import SentenceTransformer
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

# Load the document
raw_documents = TextLoader('/content/three_ghost.txt').load()

# Create a text splitter
text_splitter = CharacterTextSplitter(chunk_size=10, chunk_overlap=0)

# Split the document into chunks
documents = text_splitter.split_documents(raw_documents)

# Load a custom embedding model (using Sentence Transformers)
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")  # You can use any other model available in Hugging Face

# Compute embeddings
class CustomEmbeddings:
    def embed_query(self, text: str):
        return embedding_model.encode(text).tolist()

    def embed_documents(self, texts: list):
        return [embedding_model.encode(doc).tolist() for doc in texts]

# Use custom embeddings with Chroma
custom_embeddings = CustomEmbeddings()
db = Chroma.from_documents(documents, custom_embeddings)

# Test similarity search
query = "What happened to the ghost?"
results = db.similarity_search(query)
for result in results:
    print(result.page_content)
```

```python
# Perform a similarity search
query = "Welcome"
docs = db.similarity_search(query)
print(docs[0].page_content)

# Chat interaction with Gemini
from langchain_core.messages import HumanMessage, AIMessage

# Example of a chat interaction
chat = model.start_chat(history=[])

# Human message
human_message = HumanMessage(content="Hi! I'm Bob")
response = chat.send_message(human_message.content)
print(response.text)

# AI message
ai_message = AIMessage(content="Hello Bob! How can I assist you today?")
response = chat.send_message(ai_message.content)
print(response.text)

# Follow-up human message
human_message = HumanMessage(content="What's my name?")
response = chat.send_message(human_message.content)
print(response.text)
```
